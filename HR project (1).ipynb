# Data Pre-processing¶


```python
label_encoder = LabelEncoder()
data["department"]=label_encoder.fit_transform(data["department"])
data["region"]=label_encoder.fit_transform(data["region"])
data["education"]=label_encoder.fit_transform(data["education"])
data["gender"]=label_encoder.fit_transform(data["gender"])
data["recruitment_channel"]=label_encoder.fit_transform(data["recruitment_channel"])

scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
```

```python
inn=data[['department', 'region', 'education', 'gender',
       'recruitment_channel', 'no_of_trainings', 'age','previous_year_rating',
       'length_of_service', 'KPIs_met >80%', 'awards_won?',
       'avg_training_score']]
         
out=data[['is_promoted']]
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(inn,out,test_size=0.2,random_state=0)
```

```python
def score (pre):
    print(accuracy_score(pre,ytest)*100)
    print("----------------------------------")
    print(confusion_matrix(pre,ytest))
    print("----------------------------------")
    print(classification_report(pre,ytest))
```

# LogisticRegression

```python
logistic_model=LogisticRegression()
logistic_model.fit(xtrain,ytrain)
#without&with standered scaler 91,5
```

```python
logistic_pre=logistic_model.predict(xtest)
```

```python
score(logistic_pre)
```

# GaussianNB

```python
naive_model=GaussianNB()
naive_model.fit(xtrain,ytrain)
#without&with standered scaler 91.3
```

```python
naive_pre=naive_model.predict(xtest)
```

```python
score(naive_pre)
```

# KNN

```python
knn_model=KNeighborsClassifier(n_neighbors=5)
knn_model.fit(xtrain,ytrain)
```

```python
knn_pre=knn_model.predict(xtest)
```

```python
score(knn_pre)
```

# SVC

```python
svc_model=SVC(kernel="linear",random_state=0)
svc_model.fit(xtrain,ytrain)
```

```python
svc_pre=svc_model.predict(xtest)
```

```python
score(svc_pre)
```
